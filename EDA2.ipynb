## Data Cleaning
We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers (imputation).
However we can be smarter about this and check the average age by passenger class. For example:


```python

sns.boxplot(x='Pclass',y='Age',data=train)
```

We can see the wealthier passengers in the higher classes tend to be older, which makes sense. We'll use these average age values to impute based on Pclass for Age.

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

Now apply that function!

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
#apply used to apply custom functions. on multiple coloumns in the form of list
```

## Converting Categorical Features

We'll need to convert categorical features to dummy variables using pandas! Otherwise our machine learning algorithm won't be able to directly take in those features as inputs.

```python
train.info()
#Columns that contain categorical answers like Yes/No, Male/Female, or S/Q/C are called categorical features.
# Since machine learning algorithms can't understand text or categories directly,
#  we convert these categorical features into dummy variables using
#  One-Hot Encoding â€” where each category becomes a separate binary column (1 or 0).
```

```python
pd.get_dummies(train['Embarked'],drop_first=True).head()
# pd.get_dummies(train['Embarked'],drop_first=False).head().astype(int)


# We convert categorical variables (like "Male", "Female", "S", "C", etc.) into True/False or
#  1/0 columns so that the machine learning model can understand and process them.
# Each of these columns is called a dummy variable.
# The process of converting these categorical values into dummy variables is called One Hot Encoding (OHE).
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=False).astype(int)
embark = pd.get_dummies(train['Embarked'],drop_first=False).astype(int)

# embark = pd.get_dummies(train["Embarked"], drop_first=True)
embark.head()
sex.head()
```

```python

train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
train
```

```python
train.head()
```

```python
train = pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

Great! Our data is ready for our model!

# Building a Logistic Regression model

Let's start by splitting our data into a training set and test set (there is another test.csv file that you can play around with in case you want to use all this data for training).

## Train Test Split

```python
train.drop('Survived',axis=1).head()
```

```python
train['Survived'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1),
                                                    train['Survived'], test_size=0.30,
                                                    random_state=101)
```

## Training and Predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python

accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```
